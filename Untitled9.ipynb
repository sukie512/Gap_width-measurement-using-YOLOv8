```python
import glob
import cv2
import matplotlib.pyplot as plt

def plot_image_with_labels(img_path, lbl_path):
    img = cv2.imread(img_path)
    if img is None:
        print(f"❌ Failed to load image: {img_path}")
        return

    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    h, w, _ = img.shape

    with open(lbl_path, "r") as f:
        labels = f.readlines()

    for label in labels:
        parts = label.strip().split()
        if len(parts) < 5:
            continue  # Skip invalid labels

        class_id = int(parts[0])
        x_center, y_center, width, height = map(float, parts[1:5])

        # Convert YOLO format to pixel values
        x1 = int((x_center - width / 2) * w)
        y1 = int((y_center - height / 2) * h)
        x2 = int((x_center + width / 2) * w)
        y2 = int((y_center + height / 2) * h)

        cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 0), 2)
        cv2.putText(img, str(class_id), (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    plt.figure(figsize=(8, 8))
    plt.imshow(img)
    plt.axis("off")
    plt.show()

# Find a matching image for the first label
if label_files:
    sample_label = os.path.join(labels_path, label_files[0])

    img_extensions = [".jpg", ".jpeg", ".png"]
    sample_image = None

    for ext in img_extensions:
        img_path = os.path.join(images_path, label_files[0].replace(".txt", ext))
        if os.path.exists(img_path):
            sample_image = img_path
            break

    if sample_image:
        print("\n📷 Displaying image:", sample_image)
        plot_image_with_labels(sample_image, sample_label)
    else:
        print("❌ No matching image found for:", sample_label)
```

```python
import glob
import cv2
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm


all_images=glob.glob("/content/drive/MyDrive/extracted_frames/*.jpg")

gaps=[]

for image_path in tqdm(all_images[:100]):

  im111=cv2.imread(image_path)
  im=im111.copy()

  scale_copy=im.copy()


  w,h= im.shape[1], im.shape[0]


  im= cv2.resize(im,(640,640))


  result=model.predict(im,conf=0.3,save=True)



  boxes=result[0].boxes
  masks=result[0].masks

  scale_conf=[]
  for box in boxes:


    if box.cls[0]==1:
      scale_conf.append(box.conf[0].cpu().numpy())

  scale_index_max_conf=np.argmax(scale_conf)


  for i in range(boxes.shape[0]):



    if boxes[i].cls[0]==0:

      img1=cv2.resize(masks[i].data[0].cpu().numpy(),(w,h))
      cv2.imwrite("gap.png",img1*255)


  cou=0
  for i in range(boxes.shape[0]):
    if boxes[i].cls[0]==1:

      if cou==scale_index_max_conf:

        img1=cv2.resize(masks[i].data[0].cpu().numpy(),(w,h))
        cv2.imwrite("scale.png",img1*255)
      cou+=1

  #gap_img=cv2.imread("gap.png")
  a11=cv2.imread("gap.png")
  a11= cv2.cvtColor(a11, cv2.COLOR_BGR2GRAY)
  contours, _ = cv2.findContours(a11, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

  # Sort contours by area (largest first)
  cnt = sorted(contours, key=cv2.contourArea, reverse=True)
  peri= cv2.arcLength(cnt[0],True)
  approx = cv2.approxPolyDP(cnt[0],0.0025*peri, True)
  approx_n= np.zeros((4,2))
  approx= approx.reshape((len(approx),2))
  a= sorted(approx,key=sum,reverse=True)
  b=sorted(approx,key=sum,reverse=False)
  c= sorted(approx,key= lambda x:x[1]-x[0],reverse=True)
  d=sorted(approx,key=lambda x: x[1]-x[0],reverse=False)
  approx_n[0]=b[0]
  approx_n[3]=c[0]
  approx_n[1]=d[0]
  approx_n[2]=a[0]
  points = np.array(approx_n, np.int32)

  # Reshape to required format for polylines
  cv2.line(im111, tuple(points[0]), tuple(points[1]), (0, 255, 0), 1)  # Top edge
  cv2.line(im111, tuple(points[1]), tuple(points[2]), (0, 255, 0), 1)  # Right edge
  cv2.line(im111, tuple(points[2]), tuple(points[3]), (0, 255, 0), 1)  # Bottom edge
  cv2.line(im111, tuple(points[3]), tuple(points[0]), (0, 255, 0), 1)  # Left edge

  cv2.imwrite("gap_contour.png", im111)

  import math
  gap_in_pixels=math.dist(points[1],points[2])






  #############################################################################


  a11=cv2.imread("scale.png")
  a11= cv2.cvtColor(a11, cv2.COLOR_BGR2GRAY)
  contours, _ = cv2.findContours(a11, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

  # Sort contours by area (largest first)
  cnt = sorted(contours, key=cv2.contourArea, reverse=True)
  peri= cv2.arcLength(cnt[0],True)
  approx = cv2.approxPolyDP(cnt[0],0.0025*peri, True)
  approx_n= np.zeros((4,2))
  approx= approx.reshape((len(approx),2))
  a= sorted(approx,key=sum,reverse=True)
  b=sorted(approx,key=sum,reverse=False)
  c= sorted(approx,key= lambda x:x[1]-x[0],reverse=True)
  d=sorted(approx,key=lambda x: x[1]-x[0],reverse=False)
  approx_n[0]=b[0]
  approx_n[3]=c[0]
  approx_n[1]=d[0]
  approx_n[2]=a[0]
  points = np.array(approx_n, np.int32)

  # Reshape to required format for polylines
  cv2.line(im111, tuple(points[0]), tuple(points[1]), (0, 255, 0), 1)  # Top edge
  cv2.line(im111, tuple(points[1]), tuple(points[2]), (0, 255, 0), 1)  # Right edge
  cv2.line(im111, tuple(points[2]), tuple(points[3]), (0, 255, 0), 1)  # Bottom edge
  cv2.line(im111, tuple(points[3]), tuple(points[0]), (0, 255, 0), 1)  # Left edge

  cv2.imwrite("scale_contour.png", im111)


  height_in_pixels=math.dist(points[1],points[2])
  width_in_pixels=math.dist(points[0],points[1])


  cropped_scale = scale_copy[int(points[0][1]):int(points[0][1]+height_in_pixels), int(points[0][0]):int(points[0][0]+width_in_pixels)]



  img=cv2.cvtColor(cropped_scale,cv2.COLOR_BGR2GRAY)

  img = cv2.medianBlur(img,5)

  ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
  th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)
  th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)



  # Apply edge detection method on the image
  edges = cv2.Canny(th3, 50, 150, apertureSize=3)



  # This returns an array of r and theta values
  lines_list =[]
  lines = cv2.HoughLinesP(
              edges, # Input edge image
              1, # Distance resolution in pixels
              np.pi/180, # Angle resolution in radians
              threshold=10, # Min number of votes for valid line
              minLineLength=30, # Min allowed length of line
              maxLineGap=4 # Max allowed gap between line for joining them
              )

  copy_crop_scale=cropped_scale.copy()
  # Iterate over points

  below_points=[]
  for points in lines:
        # Extracted points nested in the list
      x1,y1,x2,y2=points[0]
      # Draw the lines joing the points
      # On the original image

      if 180*abs(math.atan((y2-y1)/(x2-x1)))/3.14>80:

        cv2.line(cropped_scale,(x1,y1),(x2,y2),(0,255,0),1)
        # Maintain a simples lookup list for points
        lines_list.append([(x1,y1),(x2,y2)])



        below_points.append([x1,y1])

  # Save the result image


  cv2.imwrite('detectedLines.png',cropped_scale)

  below_points_sorted=sorted(below_points,key=lambda x:x[0])

  point_of_inerest=[below_points_sorted[0][0],below_points_sorted[0][1]-20]

  width1=below_points_sorted[1][0]-below_points_sorted[0][0]
  box=copy_crop_scale[int(point_of_inerest[1]):int(point_of_inerest[1]+20),int(point_of_inerest[0]):int(point_of_inerest[0]+width1)]






  point_of_inerest=[below_points_sorted[1][0],below_points_sorted[1][1]-20]

  width=below_points_sorted[2][0]-below_points_sorted[1][0]
  box=copy_crop_scale[int(point_of_inerest[1]):int(point_of_inerest[1]+20),int(point_of_inerest[0]):int(point_of_inerest[0]+width)]








  scale_min_resolution_in_pixels=below_points_sorted[2][0]-below_points_sorted[1][0]+width1/3.5
  scale_min_resolution_in_mm=0.5

  mm_per_pixel=scale_min_resolution_in_mm/scale_min_resolution_in_pixels


  gap_in_mm=gap_in_pixels*mm_per_pixel
  print("gap_in_mm::::::",gap_in_mm)
  gaps.append(gap_in_mm)



import pandas as pd

pd.DataFrame(gaps).T.to_csv("gaps_TR.csv")
pd.DataFrame(gaps).to_csv("gaps.csv")









```
